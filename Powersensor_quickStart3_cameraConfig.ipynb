# Powersenosr入门教程3 - 相机设置

```python
import PowerSensor as ps
import numpy as np
import cv2
import matplotlib.pyplot as pyplot
from IPython.display import clear_output
import time
```

```python
# 这个对象用于获取摄像头数据
cam1 = ps.ImageSensor()
# 这个对象用于配置摄像头
config = ps.SensorConfig()
```

## 正常环境-弱光环境设置

### 弱光环境下弱光模式的拍摄

```python
config.set_exposure(mode=ps.SensorPara.EXPOSURE_MODE_AUTO, desired_lumin=48, max_gain=64)
config.set_light_mode(ps.SensorPara.Light_Dark)
for i in range(50):
    start = time.time()        # 记录开始时间
    clear_output(wait=True)    # 清除图片，在同一位置显示，不使用会打印多张图片
    imgMat = cam1.read_img_ori()       # 读入图像
    
    # 缩小图像为320x240尺寸
    origin = cv2.resize(imgMat, (320,240))
    
    ps.CommonFunction.show_img_jupyter(imgMat)# 打印用于差分的两张图片
    end = time.time()        # 记录结束时间
    print(end - start)
    time.sleep(0.1)
```

## 手动曝光-自动曝光设置

### 自动曝光需要设置期望亮度（流明）、最大增益

```python
# 自动曝光，亮度=58，最大增益64
config.set_exposure(mode=ps.SensorPara.EXPOSURE_MODE_AUTO, desired_lumin=58, max_gain=64)
for i in range(30):
    start = time.time()        # 记录开始时间
    clear_output(wait=True)    # 清除图片，在同一位置显示，不使用会打印多张图片
    imgMat = cam1.read_img_ori()       # 读入图像
    
    # 缩小图像为320x240尺寸
    origin = cv2.resize(imgMat, (320,240))
    
    ps.CommonFunction.show_img_jupyter(imgMat)# 打印用于差分的两张图片
    end = time.time()        # 记录结束时间
    print('0-58-64', end - start)
    time.sleep(0.1)
    
```
