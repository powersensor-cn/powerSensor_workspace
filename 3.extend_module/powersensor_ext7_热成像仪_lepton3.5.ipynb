```python
import PowerSensor as ps
import time
import numpy as np
import LeptonHelper
```

# 初始化接口 
Lepton 3.5使用I2C配置参数，使用SPI传输图像，因此需要初始化这两个接口

```python
# 初始化i2c，运行一次即可
i2c = ps.I2cPort()
spi = ps.SpiPort()
spi.set_clock(40000000, 1, 1)
# 调用LeptonHelper辅助类
lepton = LeptonHelper.Lapton_Helper(i2c, spi)
```

# 寄存器读取测试

```python
x = lepton.read_reg(0x02)
print('state reg:', x[0], x[1])
lepton.lepton_cmd(LeptonHelper.SYS, 0x28 >> 2, LeptonHelper.GET)
res = lepton.lepton_read_data()
print('serial:',str(res))
```
